## 此时点源坐标$(0,56,0)cm$,距离身体表面50cm

```c++
TFile* ipf=new TFile("B1.root");
TCanvas* c1=new TCanvas;
```

```c++
TTree *ipt=(TTree*)ipf->Get("B1");
```

```c++
ipt->Draw("EventID");
c1->Draw();
```

```c++
TCanvas* c3=new TCanvas;
c3->Divide(1,2);
```

```c++
c3->cd(1);
ipt->Draw("Energy_head");
c3->cd(2);
ipt->Draw("Energy_body");
c3->Draw();
```

## 在头部和身体沉积的能量都非常小，在图上无法看出明显的趋势

```c++
TCanvas* c2=new TCanvas;
c2->Divide(1,3);
```

```c++
c2->cd(1);
ipt->Draw("px");
c2->cd(2);
ipt->Draw("py");
c2->cd(3);
ipt->Draw("pz");
c2->Draw();
```

## 所有事件起始方向在XYZ的分布都是均匀分布（点源各向同性）

```c++
c2->cd(1);
ipt->Draw("head_x");
c2->cd(2);
ipt->Draw("head_y");
c2->cd(3);
ipt->Draw("head_z");
c2->Draw();
```

## 由于头离点源较远，能量沉积重心的分布没有太大趋势

```c++
c2->cd(1);
ipt->Draw("body_x");
c2->cd(2);
ipt->Draw("body_y");
c2->cd(3);
ipt->Draw("body_z");
c2->Draw();
```

## 明显可以看到在身体Y方向上的沉积能量重心在正面靠近点源处较多，如果事件数足够多，会有明显的剂量剪成区
